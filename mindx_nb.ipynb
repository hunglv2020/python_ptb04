```python
you = "Mindx"
place = "Nguyen Huu Tho"
age = 1000
#Cách 1: Truyền lần lượt giá trị trong format vào string {}.
print("Hello {}. I'm from {}. I'm {} years old.".format(you,place,age))
#Cách 2: Đảo thứ tự giá trị trong format truyền vào string {}.
print("Hello {2}. I'm from {1}. I'm {0} years old.".format(you,place,age))
#Cách 3: f-format, chỉ định đích dánh biến-giá trị.
print(f"Hello {you}. I'm from {place}. I'm {age} years old.")
#Cách 4: Đố bít
```

Example 0: Xây dựng chương trình nhập vào số nguyên dương a, b. Tính tổng tất cả các số nguyên [a,b] và in ra màn hình.

Example 1: Xây dựng chương trình nhập vào một số nguyên dương n. Hãy tính tổng các số chẵn từ 1 đến n bằng cách sử dụng vòng lặp for và in ra màn hình kết quả.

```python
a = int(input("Nhap so nguyen a: "))
b = int(input("Nhap so nguyen b: "))
sum = 0
for i in range(a,b):
    if i%2 == 1:
        sum = sum + i
print(sum)
```

Example 2: Xây dựng chương trình nhập vào một số nguyên dương n. Hãy xử lí và in ra màn hình các số chẵn từ 1 đến n.

```python
n = int(input("Nhap so nguyen duong n: "))
for i in range(1,n):
    if i%2 == 0 :
        print(i)
```

Example 3: Xây dựng chương trình tính giai thừa của một số nguyên dương n. Xử lí và in ra màn hình kết quả.

```python
n = int(input("Nhap so nguyen duong n: ")) # Gia su nhap vao so 4
giaithua = 1
for i in range(1,n+1): #range(0,5)
    giaithua = giaithua * i
print(giaithua)
```
