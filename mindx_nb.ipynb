<h4>2. Làm việc với String format</h4>

```python
you = "Mindx"
place = "Nguyen Huu Tho"
age = 1000
#Cách 1: Truyền lần lượt giá trị trong format vào string {}.
print("Hello {}. I'm from {}. I'm {} years old.".format(you,place,age))
#Cách 2: Đảo thứ tự giá trị trong format truyền vào string {}.
print("Hello {2}. I'm from {1}. I'm {0} years old.".format(you,place,age))
#Cách 3: f-format, chỉ định đích dánh biến-giá trị.
print(f"Hello {you}. I'm from {place}. I'm {age} years old.")
#Cách 4: Đố bít
```

Example 2: Write a program to accept from 1 to 7 and display name of the day 1 for Sunday, 2 for Monday, ... <br>
Bài tâp 2: Viết một chương trình để người dùng nhập vào một số từ 1 đến 7. Kiểm tra số đó, nếu là 1 thì in ra là "Sunday", nếu là số 2 in ra là "Monday", ... 

```python
#Solution for Example 2
n = int(input("Nhap mot so nguyen tu 1 den 7"))
if n == 1:
    print("Sunday")
elif n == 2:
    print("Monday")
elif n == 3: 
    print("Tuesday")
else:
    print("So ban vua nhap khong chinh xac")
```
